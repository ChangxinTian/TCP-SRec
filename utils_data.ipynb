```python
import pandas as pd
```

```python
import time

def timestamp_convert_localdate(timestamp,time_format="%Y/%m/%d %H:%M:%S"):
    timeArray = time.localtime(timestamp)
    print(timeArray)
    styleTime = time.strftime(str(time_format), timeArray)
    return styleTime
```

```python
# data_path = "/home/tianchangxin/22CIKM/RecBole/dataset/amazon-beauty/amazon-beauty.inter"
# time_field = 'timestamp:float'

data_path = "/home/tianchangxin/22CIKM/TPRec/dataset/yelp/yelp.inter"
time_field = 'date:float'
```

```python
data_df = pd.read_csv(data_path, sep='\t')
print(data_df.shape)
data_df.head()
```

```python
data_df = data_df.sort_values(by=['user_id:token', time_field])
data_df.head()
```

```python
data_df_5 = data_df
```

```python
print(data_df_5.shape)
data_df_5['tm_hour:float'] = data_df_5[time_field].map(lambda x: time.localtime(x).tm_hour)
data_df_5['tm_wday:float'] = data_df_5[time_field].map(lambda x: time.localtime(x).tm_wday)
data_df_5['tm_mon:float'] = data_df_5[time_field].map(lambda x: time.localtime(x).tm_mon)
data_df_5['tm_year:float'] = data_df_5[time_field].map(lambda x: time.localtime(x).tm_year)
data_df_5.head()
```

```python
from tqdm import tqdm

session_id = []
pre_user, pre_time, pre_session = '', 0, 0
for index, row in tqdm(data_df_5.iterrows()):
    cur_user = row['user_id:token']
    cur_time = float(row[time_field])
    if cur_user == pre_user:
        if cur_time - pre_time > 3600:
            cur_session += 1
    else:
        cur_session = 0
    session_id.append(str(cur_user)+'-'+str(cur_session))
    pre_user, pre_time, pre_session = cur_user, cur_time, cur_session

data_df_5.insert(5, column='session_id:token', value=session_id)
data_df_5.tail()
```

```python
data_df_5.to_csv(data_path, sep='\t', index=False)
```
